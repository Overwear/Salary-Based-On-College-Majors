```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read data set
recent_grads = pd.read_csv("recent-grads.csv")

#Taking a look at the data
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
```

## Next Up..

I will be removing rows with missing data so that my pandas functions will not break!

```python
#Removing rows with missing data to ensure all functions work correctly
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
num_rows_removed = raw_data_count - cleaned_data_count
print("The number of rows dropped due to missing values is: " + str(num_rows_removed))
```

## Comments

Nice! Only 1 row had missing data.

## Next Up..

I will use Dataframe.plot to visually inspect the data with scatter plots


```python
#Generate scatter plot of Sample_size and Median
ax1 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median Salary vs Sample Size', xlim=(0, 4500))

#Generate scatter plot of Sample_size and Unemployment_rate
ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment Rate vs Sample Size', xlim=(-10,4500))

#Generate scatter plot of Full_time and Median
ax3 = recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Number of Full Time Workers Vs Median Salary', xlim=(-10,275000))

#Generate scatter plot of ShareWomen and Unemployment_rate
ax4 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Women\'s Percentage vs Unemployment Rate')

#Generate scatter plot of Men and Median
ax5 = recent_grads.plot(x='Men', y='Median', kind='scatter', title='Male Graduates vs Median Salary', xlim=(-10,200000))

#Generate scatter plot of Women and Median
ax6 = recent_grads.plot(x='Women', y='Median', kind='scatter', title='Female Graduates vs Median Salary', xlim=(-10,200000))

#Generate scatter plot of ShareWomen and Median
ax7 = recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='Percentage of Woman in Major vs Median Salary')
```

## Comments

<b>Question 1)</b> Do students in more popular majors make more money?
<br><b>Answer 1)</b> To answer this question I will be looking at the Median Salary vs Sample Size chart. From the chart I can see that there is a weak negative linear relationship between sample size and the median salary. The trend tends to say that the higher the sample size (the more popular majors) tend to have a lower median salary. 

<b>Question 2)</b> Do students that majored in subjects that were majority females make more money?
<br><b>Answer 2)</b> To answer this question I will be referring to the Percentage of Women in Major Vs Median salary scatter plot. There is a strong negative linear relationship between the percentage of women in the major and the median salary. 

<b>Question 3)</b> Is there a link between the number of full-time employees and median salary?
<br><b>Answer 3)</b>There doesn't seem to be an apparent relationship between the number of full time employees and median salary from the scatter plot.

## Next Up..

I will use a scatter matrix plot which combines both scatter plots and histograms into one grid of plots. This will help visualize potential relationships and distributions all at once. 

```python
#import the scatter matrix module
from pandas.plotting import scatter_matrix

#Explore the sample size vs median using scatter matrix
scatter_matrix(recent_grads[["Sample_size", "Median"]], figsize=(10,10))
```

```python
#Exploring 3x3 scatter matrix using Sample_size, Median and Unemployment_rat
scatter_matrix(recent_grads[["Sample_size", "Median", "Unemployment_rate"]], figsize=(10,10))
```

## Comments

From these scatter matrices I can see that the median salary of around 40k had the highest sample sizes as well.

The unemployment rate tends to be higher amoung the lower salary range as compared to the higher end of the range.
